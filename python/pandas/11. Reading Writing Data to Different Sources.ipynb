# Reading/Writing Data to Different Sources - Excel, JSON, SQL, Etc

```python
import pandas as pd
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent')
schema_df = pd.read_csv('data/survey_results_schema.csv', index_col='Column')
```

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)

df.head()
```

```python
filt = (df['Country'] == 'Indonesia')
indo_df = df.loc[filt]
indo_df.head()
```

#### export into csv file

```python
indo_df.to_csv('data/indonesia.csv')
```

#### export into tsv file

```python
indo_df.to_csv('data/indonesia.tsv', sep='\t')
```

#### export into excel file

```python
indo_df.to_excel('data/indonesia.xlsx')
```

#### read excel file

```python
excel = pd.read_excel('data/indonesia.xlsx')
```

```python
excel.head()
```

#### export to json

```python
indo_df.to_json('data/indonesia.json', orient='records', lines=True)
```
