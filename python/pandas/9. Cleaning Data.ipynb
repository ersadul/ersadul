# Cleaning Data - Casting Datatypes and Handling Missing Values

```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['steve', 'jane', 'john', 'adam', np.nan, None, 'NA'],
    'last': ['jobs', 'doe', 'doe', 'doe', np.nan, np.nan, 'Missing'],
    'email': ['stevejob@email.com', 'janedoe@email.com', 'johndoe@email.com', None, np.nan, 'anon@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

add replace method to change 'NA' and 'Missing' value with NaN

```python
df = pd.DataFrame(people)

df.replace(['NA', 'Missing'], np.nan, inplace=True)
df
```

```python
df.dropna()
```

Default arguments of dropna

axis: it can be 'index' or 'columns'
 * 0, or ‘index’ : Drop rows which contain missing values.
 * 1, or ‘columns’ : Drop columns which contain missing value.
 
how: determine if row or column will dropped
 * ‘any’ : drop an row or column if there is null value
 * ‘all’ : drop an row or column if all of row/column is null value
 
subset: remove null value in selected columns

```python
df.dropna(axis='index', how='all', subset=['first', 'age'])
```

```python
df.isna()
```

```python
df['age'].fillna(0)
```

#### Casting values

```python
df.dtypes
```

```python
df['age'] = df['age'].astype(float)
```

```python
df.dtypes
```

```python
df['age'].mean()
```
